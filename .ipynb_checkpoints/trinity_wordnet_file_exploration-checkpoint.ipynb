This notebook builds on our class examples with NLTK, part-of-speech tagging, and WordNet.

## What I wanted to do
- read a small collection of `.txt` files from my own folder
- tokenize and tag the words
- look at the most common nouns, verbs, and adjectives
- use WordNet to compare how many synsets some words have
- save a couple of results to output files instead of only printing in the notebook

I left comments in the code cells so I can keep working on it in class.

```python
from pathlib import Path
from collections import Counter
import csv
import nltk
from nltk import word_tokenize, pos_tag
from nltk.corpus import wordnet as wn

# Download NLTK resources if needed.
# If these are already installed in your .venv, this cell should finish quickly.
downloads = [
    "punkt",
    "averaged_perceptron_tagger",
    "wordnet",
    "omw-1.4",
]

for pkg in downloads:
    nltk.download(pkg)
```

## Set up my folders

This notebook assumes I am working inside my own project directory.

I should put my text files in a folder such as:
- `text-files/`
or
- `python-nlp/text-files/`

The code below looks for a local `text-files` folder first. If I rename that folder, I just need to change one line.

```python
# Change this path if your text files live somewhere else.
TEXT_DIR = Path("text-files")
OUTPUT_DIR = Path("output-files")

OUTPUT_DIR.mkdir(exist_ok=True)

if not TEXT_DIR.exists():
    print("TEXT_DIR was not found yet.")
    print("Create a folder called 'text-files' and put your .txt files in it,")
    print("or edit the TEXT_DIR variable in this cell.")
else:
    print("Using text directory:", TEXT_DIR.resolve())
    print("Output will be saved to:", OUTPUT_DIR.resolve())
```

```python
text_files = sorted(TEXT_DIR.glob("*.txt")) if TEXT_DIR.exists() else []

print("Files found:", len(text_files))
for f in text_files:
    print("-", f.name)
```

## Read my file collection

Here I loop through all `.txt` files in my folder and combine them into one big string for analysis.

```python
def read_collection(file_list):
    corpus_parts = []
    for file_path in file_list:
        try:
            text = file_path.read_text(encoding="utf-8")
        except UnicodeDecodeError:
            text = file_path.read_text(encoding="latin-1")
        corpus_parts.append(text)
    return "\n\n".join(corpus_parts)

full_text = read_collection(text_files) if text_files else ""
print("Character count:", len(full_text))
print(full_text[:500] if full_text else "No text loaded yet.")
```

## Tokenize and tag the words

This is where I start exploring part-of-speech patterns in my files.

```python
tokens = word_tokenize(full_text) if full_text else []
alpha_tokens = [t.lower() for t in tokens if any(ch.isalpha() for ch in t)]
tagged_tokens = pos_tag(alpha_tokens) if alpha_tokens else []

print("Total tokens:", len(tokens))
print("Alphabetic-ish tokens:", len(alpha_tokens))
print("Tagged tokens:", len(tagged_tokens))
print(tagged_tokens[:25])
```

```python
# Separate out a few POS categories
nouns = [word for word, tag in tagged_tokens if tag.startswith("NN")]
verbs = [word for word, tag in tagged_tokens if tag.startswith("VB")]
adjectives = [word for word, tag in tagged_tokens if tag.startswith("JJ")]

noun_counts = Counter(nouns)
verb_counts = Counter(verbs)
adj_counts = Counter(adjectives)

print("Top nouns:", noun_counts.most_common(10))
print("Top verbs:", verb_counts.most_common(10))
print("Top adjectives:", adj_counts.most_common(10))
```

## Use WordNet

Now I want to compare some frequent words with WordNet and see how many synsets they have.

A bigger number of synsets can suggest a word is more flexible or ambiguous in meaning.

```python
def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith("J"):
        return wn.ADJ
    if treebank_tag.startswith("V"):
        return wn.VERB
    if treebank_tag.startswith("N"):
        return wn.NOUN
    if treebank_tag.startswith("R"):
        return wn.ADV
    return None

def synset_count(word, pos_tag_label=None):
    wn_pos = get_wordnet_pos(pos_tag_label) if pos_tag_label else None
    synsets = wn.synsets(word, pos=wn_pos) if wn_pos else wn.synsets(word)
    return len(synsets), synsets

top_verbs_with_tags = [(w, tag) for w, tag in tagged_tokens if tag.startswith("VB")]
seen = set()
unique_top_verbs = []
for w, tag in top_verbs_with_tags:
    if w not in seen:
        unique_top_verbs.append((w, tag))
        seen.add(w)
    if len(unique_top_verbs) >= 15:
        break

verb_synset_rows = []
for word, tag in unique_top_verbs:
    count, synsets = synset_count(word, tag)
    example_names = [s.name() for s in synsets[:3]]
    verb_synset_rows.append((word, tag, count, example_names))

verb_synset_rows
```

## Save results to files

This part matters for the assignment because it creates files outside the console.

I made:
- a text report
- a CSV file for the top words by category

```python
def write_text_file(out_path, text):
    out_path.write_text(text, encoding="utf-8")
    print(f"Wrote {out_path}")

def write_csv_file(out_path, rows, header):
    with out_path.open("w", encoding="utf-8", newline="") as f:
        writer = csv.writer(f)
        writer.writerow(header)
        writer.writerows(rows)
    print(f"Wrote {out_path}")

summary_lines = []
summary_lines.append("NLTK + WordNet exploration summary\n")
summary_lines.append(f"Files analyzed: {len(text_files)}")
summary_lines.append(f"Total tokens: {len(tokens)}")
summary_lines.append(f"Total tagged tokens: {len(tagged_tokens)}\n")

summary_lines.append("Top 10 nouns:")
for word, count in noun_counts.most_common(10):
    summary_lines.append(f"- {word}: {count}")

summary_lines.append("\nTop 10 verbs:")
for word, count in verb_counts.most_common(10):
    summary_lines.append(f"- {word}: {count}")

summary_lines.append("\nTop 10 adjectives:")
for word, count in adj_counts.most_common(10):
    summary_lines.append(f"- {word}: {count}")

summary_lines.append("\nVerb synset comparison:")
for word, tag, count, syn_examples in verb_synset_rows:
    summary_lines.append(f"- {word} ({tag}) -> {count} synsets | examples: {', '.join(syn_examples)}")

write_text_file(OUTPUT_DIR / "nltk_wordnet_summary.txt", "\n".join(summary_lines))

top_word_rows = []
for label, counter in [("noun", noun_counts), ("verb", verb_counts), ("adjective", adj_counts)]:
    for word, count in counter.most_common(15):
        top_word_rows.append((label, word, count))

write_csv_file(
    OUTPUT_DIR / "top_words_by_pos.csv",
    top_word_rows,
    ["category", "word", "count"]
)

write_csv_file(
    OUTPUT_DIR / "verb_synset_counts.csv",
    [(word, tag, count, "; ".join(examples)) for word, tag, count, examples in verb_synset_rows],
    ["word", "pos_tag", "synset_count", "example_synsets"]
)
```

## Quick reflection

One thing I noticed is that POS tagging helps me separate the text into patterns instead of just looking at a giant wall of words. WordNet also makes it easier to see that some of the verbs in my files can carry a lot of different meanings, which is interesting when I compare dialogue-heavy or repetitive text files.

If I get stuck later, I want to keep experimenting with:
- concordances for repeated words
- collocations or frequent word pairs
- comparing one file against another instead of combining everything together

```python
# Optional next step:
# compare files one by one instead of only making one combined corpus

per_file_report = []

for file_path in text_files:
    try:
        single_text = file_path.read_text(encoding="utf-8")
    except UnicodeDecodeError:
        single_text = file_path.read_text(encoding="latin-1")

    single_tokens = [t.lower() for t in word_tokenize(single_text) if any(ch.isalpha() for ch in t)]
    single_tagged = pos_tag(single_tokens)

    single_verbs = [word for word, tag in single_tagged if tag.startswith("VB")]
    single_verbs_count = Counter(single_verbs).most_common(10)

    per_file_report.append(f"\nFILE: {file_path.name}")
    for word, count in single_verbs_count:
        per_file_report.append(f"{word}: {count}")

if per_file_report:
    write_text_file(OUTPUT_DIR / "per_file_top_verbs.txt", "\n".join(per_file_report))
else:
    print("No per-file report yet because no text files were found.")
```
