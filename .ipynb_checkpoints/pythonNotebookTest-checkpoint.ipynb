3b. Prepare a code block that depends on running the previous code block and then reads from each directory of files and calculates what's needed to indicate **lexical diversity**.
    Running the cell that you prepare should reveal the lexical diversity of each set of files. *(3 pts)*

```python
def get_tokens(folder):
    tokens = []
    for file in os.listdir(folder):
        if file.endswith(".txt"):
            with open(os.path.join(folder, file), encoding="utf-8") as f:
                text = f.read()
                tokens.extend(word_tokenize(text))
    return tokens

def lexical_diversity(tokens):
    return len(set(tokens)) / len(tokens)

tokens1 = get_tokens("AT")
tokens2 = get_tokens("RS")

print("Adventure Time:", lexical_diversity(tokens1))
print("Regular Show:", lexical_diversity(tokens2))
```

4. Comment what the results are showing you about your texts in your own words. From what you can see, which set has the highest lexical diversity, and what does that mean about the language that set of files, compared to the other set? *(3 pts)* 

The results show that the Adventure Time texts have a higher lexical diversity (0.1458) compared to the Regular Show texts (0.1200). This means that Adventure Time uses a wider variety of words, while Regular Show repeats words more often. A higher lexical diversity suggests more varied or creative language, while a lower one suggests more repetition in vocabulary.

5. Now, write a code block that depends on variables you set for your file directories in the previous code block(s). This codeblock should just select a set of tokens from your text based on their part of speech, either adjectives or verbs.  *(5 pts)* 

```python
# Select 
def get_verbs(tokens):
    tagged = pos_tag(tokens)  
    verbs = [word for word, tag in tagged if tag.startswith("VB")]  # keep only verbs
    return verbs

# Apply 
verbs1 = get_verbs(tokens1)
verbs2 = get_verbs(tokens2)

# Print
print("Sample Adventure Time verbs:", verbs1[:20])
print("Sample Regular Show verbs:", verbs2[:20])
```

6. Develop the code in the previous block by copying it into the next code block, and here:
   * Develop code to find out about the top 15 most ambiguous words in your adjective or verb token list, and
   * Write your output to a neatly formatted text file for reading. The text output file should provide a clear heading to provide a title for each corpus you're comparing. *(5 pts)*

```python

def get_ambiguity(word):
    return len(wordnet.synsets(word))

def top_ambiguous(words):
    scores = {}
    for word in words:
        scores[word] = get_ambiguity(word)
    sorted_words = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    return sorted_words[:15]

top1 = top_ambiguous(verbs1)
top2 = top_ambiguous(verbs2)


with open("ambiguity_results.txt", "w") as f:
    f.write("Adventure Time - Top Ambiguous Verbs:\n")
    for word, score in top1:
        f.write(f"{word}: {score}\n")

    f.write("\nRegular Show - Top Ambiguous Verbs:\n")
    for word, score in top2:
        f.write(f"{word}: {score}\n")

print("File written successfully")
```
